# Data Exploration
This notebook explores the FER-2013 dataset.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import cv2

# Ensure we are in the correct directory
import sys
sys.path.append('../')
from src.utils import EMOTION_LABELS
```

```python
# Load data
X_train = np.load('../data/processed/X_train.npy')
y_train = np.load('../data/processed/y_train.npy')
print(f'Training shape: {X_train.shape}, Labels shape: {y_train.shape}')
```

## Class Distribution
Identifying class imbalance.

```python
labels = np.argmax(y_train, axis=1)
counts = Counter(labels)
plt.figure(figsize=(10,6))
sns.barplot(x=[EMOTION_LABELS[k] for k in counts.keys()], y=list(counts.values()))
plt.title('Emotion Class Distribution (Training Set)')
plt.show()
```

## Sample Images
Displaying sample images for each emotion.

```python
plt.figure(figsize=(15, 10))
for i in range(7):
    idx = np.where(labels == i)[0][0]
    plt.subplot(2, 4, i+1)
    plt.imshow(X_train[idx].reshape(48, 48), cmap='gray')
    plt.title(EMOTION_LABELS[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
